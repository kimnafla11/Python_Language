```python
ans = [[]for i in range(9)]
for i in range(1,10):
    for j in range(1,10):
        ans[j-1].append('%d*%d=%d'%(i,j,i*j))
print(ans)
```

```python
ans = [[]for i in range(9)]
for i in range(9,0,-1):
    for j in range(1,10):
        ans[j-1].append('%d*%d=%d'%(i,j,i*j))
print(ans)
```

```python
ans = [[]for i in range(9)]
for i in range(1,10):
    for j in range(1,10):
        if(i!=2 and i!=5):
            ans[j-1].append('%d*%d=%d'%(i,j,i*j))
print(ans)
```

```python
a = 12
b = 6
for i in range(1,b+1):
    if(a%i==0 and b%i==0):
        print(i)
```

```python
for i in range(b,0,-1):
    if(a%i==0 and b%i==0):
        print(i)
        break
    
```

```python
for i in range(1,a*b+1):
    if(i%a==0 and i%b==0):
        print(i)
        break
```

```python
n_p = list([i for i in range(2,101) for j in range(2,i) if i%j==0])
for i,j in enumerate(n_p):
    print('%d,%d'%(i+1,j))
```

```python
n_p = list(set([i for i in range(2,101) for j in range(2,i) if(i%j==0)]))
for i,j in enumerate(n_p):
    print('%d,%d'%(i+1,j))
```
