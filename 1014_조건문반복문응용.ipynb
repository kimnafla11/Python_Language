```python
# 기본1. 1~100까지 숫자중 홀수만 출력
for i in range(1,101,+2):
    print(i)
    
for i in range(1,101):
    if i % 2 == 1:
        print(i)
```

```python
# 기본 2. 1단부터 9단까지 순서대로 구구단 출력
for i in range(1,10):
    for j in range(1,10):
        print('%d x %d = %d' %(i,j,i*j))
    print('')
```

```python
# 기본 3. 9단부터 1단까지 순서로 구구단 (앞의 숫자만 역순, 뒤의 숫자는 순서대로)

for i in range(9,0,-1): ##range(9,0,-1) 우와 이거 졸라 헷갈림★★★★★★
    for j in range(1,10):
        print('%d x %d = %d' %(i,j,i*j))
```

```python
# 기본 4. 1단부터 9단까지 구구단 중 2단, 5단을 제외하고 출력(앞자리에 오는 숫자만을 기준으로 제외)
for i in range(1,10):
    if (i != 2 and i !=5):
        for j in range(1,10):
            print('%d x %d = %d' %(i,j,i*j))
```

```python
# 기본 5. "*"하나와 반복문을 활용하여 출력

# 방법1 줄수 i , 별표 수 j
for i in range(1,6): #5줄
    for j in range(1,i+1):
        print('*', end = '')
    print('')
    
# 방법2 파이썬은 문자열 연산이 가능하다
for i in range(1,6):
    print('*'*i)
```

```python
# 기본 6.
for i in range(5,0,-1): #5줄
    for j in range(1,i+1):
        print('*', end = '')
    print('')
```
